### **Netflix Movies and TV Shows:**

The dataset contains information about movies and TV shows available on Netflix as of 2021. Your task is to perform the following actions based on the data:

1. Identify missing values: Check if there are any missing values in the dataset and decide how to handle them. For example, if the "director" column has missing values, we can either delete those rows or fill in the missing values with "unknown".

2. Data Cleaning: Clean the dataset to remove any duplicates, inconsistencies or irrelevant columns. For example, we can remove the "show_id" and "description" columns as they may not be relevant for our analysis.

3. Data Exploration: Use Matplotlib to create visualizations that explore the data in meaningful ways. For example, we can create a bar chart to show the number of movies vs TV shows available on Netflix, or a line chart to show the trend in the number of new titles added each year.

4. Analyze Genres: Analyze the genres of movies and TV shows available on Netflix using bar charts or pie charts. For example, we can create a bar chart to show the top 10 most popular genres on Netflix.

netflix (1).png

```python
# Write your code here
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/netflix_titles.csv')

# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Drop irrelevant columns
df = df.drop(['show_id', 'description'], axis=1)

# Remove duplicates
df = df.drop_duplicates()

# Check for inconsistencies and correct if necessary
# Example: Convert column values to lowercase for consistency
df['director'] = df['director'].str.lower()
```

```python
import matplotlib.pyplot as plt

# Count the number of movies and TV shows
type_counts = df['type'].value_counts()

# Create a bar chart to visualize the counts
plt.bar(type_counts.index, type_counts.values)
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Number of Movies vs TV Shows on Netflix')
plt.show()
```

```python
# Extract the genres column
genres = df['listed_in']

# Count the occurrences of each genre
genre_counts = genres.str.split(', ').explode().value_counts()

# Select the top 10 genres
top_10_genres = genre_counts.head(10)

# Create a bar chart for the top 10 genres
plt.bar(top_10_genres.index, top_10_genres.values)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Top 10 Genres on Netflix')
plt.xticks(rotation=45)
plt.show()

# Alternatively, create a pie chart for the top 10 genres
plt.pie(top_10_genres.values, labels=top_10_genres.index, autopct='%1.1f%%')
plt.title('Top 10 Genres on Netflix')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Extract the ratings column
ratings = df['rating']

# Remove any missing or unknown ratings
ratings = ratings.dropna()

# Create a histogram to visualize the rating distribution
plt.hist(ratings, bins=10)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Rating Distribution on Netflix')
plt.show()
```

Overall Impression:

Netflix offers a diverse collection of movies and TV shows, with a strong emphasis on movies.
The top genres on Netflix provide insights into the content preferences of its users, highlighting the genres that receive the most attention and viewership.
The rating distribution suggests that Netflix offers content spanning a wide range of ratings, indicating a variety of quality levels and viewer opinions.

```python
# Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a pandas DataFrame
df = pd.read_csv('/content/drive/MyDrive/Himanshi/US_Accidents_Dec21_updated.csv')

# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Count the number of accidents per state
top_10_states = df['State'].value_counts().head(10)

# Create a bar chart to visualize the data
plt.bar(top_10_states.index, top_10_states.values)
plt.xlabel('State')
plt.ylabel('Number of Accidents')
plt.title('Top 10 States with the Highest Number of Accidents')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a scatter plot of longitude and latitude
plt.scatter(df['Start_Lng'], df['Start_Lat'], alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Accident Locations')
plt.show()
```

```python
# Count the number of accidents per city
top_10_cities = df['City'].value_counts().head(10)

# Create a horizontal bar chart to visualize the data
plt.barh(top_10_cities.index, top_10_cities.values)
plt.xlabel('Number of Accidents')
plt.ylabel('City')
plt.title('Top 10 Cities with the Highest Number of Accidents')
plt.show()
```

```python
# Count the distribution of accidents by severity
severity_counts = df['Severity'].value_counts()

# Create a pie chart to visualize the data
plt.pie(severity_counts.values, labels=severity_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Accidents by Severity')
plt.show()
```

```python
# Convert the Start_Time column to datetime type
df['Start_Time'] = pd.to_datetime(df['Start_Time'])

# Extract the year from the Start_Time column
df['Year'] = df['Start_Time'].dt.year

# Count the number of accidents per year
accidents_per_year = df['Year'].value_counts().sort_index()

# Create a line chart to visualize the trend
plt.plot(accidents_per_year.index, accidents_per_year.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.title('Trend of Accidents over Time')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset from CSV
titanic_df = pd.read_csv('/content/drive/MyDrive/Himanshi/titanic.csv')

# Check for missing values in the Age column
missing_values = titanic_df['Age'].isnull().sum()
print("Missing values in Age column:", missing_values)

# Fill in missing values in the Age column using mean imputation
titanic_df['Age'].fillna(titanic_df['Age'].mean(), inplace=True)

# Check if there are inconsistencies in the formatting of the gender column
gender_counts = titanic_df['Sex'].value_counts()
print("Gender counts:\n", gender_counts)

# Convert the data type of the Age column to numeric
titanic_df['Age'] = pd.to_numeric(titanic_df['Age'], errors='coerce')

# Identify outliers or inconsistencies using box plots
sns.boxplot(data=titanic_df, x='Sex', y='Age', hue='Survived')
plt.title('Age Distribution by Gender and Survival')
plt.show()

# Create a bar chart to visualize the number of survivors by gender
sns.countplot(data=titanic_df, x='Sex', hue='Survived')
plt.title('Number of Survivors by Gender')
plt.show()

# Create a heatmap to visualize the correlation between variables
correlation_matrix = titanic_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset from CSV
iris_df = pd.read_csv('/content/drive/MyDrive/Himanshi/Iris.csv')

# Explore the data
print("Head of the dataset:")
print(iris_df.head())
print("\nInformation about the dataset:")
print(iris_df.info())
print("\nSummary statistics of the dataset:")
print(iris_df.describe())

# Data visualization
# Scatter plot
sns.scatterplot(data=iris_df, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.title('Sepal Length vs Sepal Width')
plt.show()

# Box plot
sns.boxplot(data=iris_df, x='Species', y='PetalLengthCm')
plt.title('Petal Length by Species')
plt.show()

# Violin plot
sns.violinplot(data=iris_df, x='Species', y='PetalWidthCm')
plt.title('Petal Width by Species')
plt.show()

# Pair plot
sns.pairplot(data=iris_df, hue='Species')
plt.title('Pairwise Relationships')
plt.show()
```

**Interpretation:**
- The scatterplot shows the relationship between sepal length and sepal width, with different species distinguished by color. We can observe patterns and clusters of data points.
- The boxplot and violinplot show the distribution of petal length and petal width, respectively, for each species. We can compare the distributions and identify any differences between the species.
- The pairplot provides a matrix of scatterplots for each pair of variables, with different species distinguished by color. We can examine the relationships between variables and how they vary across species.
